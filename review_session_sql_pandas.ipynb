# SQL

## Loading a database
1. in your terminal, navigate to the folder containing the databasename.sql file
2. psql to enter the postgres terminal
3. CREATE DATABASE databasename; where databasename is the name of the database
4. \q to exit the postgres terminal
5. psql databasename < databasename.sql to load the database
6. psql databasename to log in to the database and start SQLing!!

### Let's just use the CITY table. Write a query for the following:

##### Quit psql:
\q

##### Show tables:
\d

##### Show first 10 rows of a table
SELECT * FROM city LIMIT 10;

##### What is the total world population?
SELECT SUM(population) from city;

##### What is the population of each country?
SELECT countrycode, sum(population) from city group by countrycode limit 10;

##### What is the population of each country, ordered highest to lowest population?
SELECT countrycode, sum(population) from city group by countrycode order by 2 DESC limit 10;

##### What is the population of each country, ordered alphabetically?
SELECT countrycode, sum(population) from city group by 1 order by 1 limit 10;

```python
import pandas as pd
```

```python
df = pd.read_csv('hospital-costs.csv')
```

```python
df.info()
```

```python
df.head()
```

```python
# Add a new column 'Total Cost'
df['Total_Cost'] = df['Discharges'] * df['Mean Cost']
df.head()
```

```python
# Group by facility name and get the total cost for each hospital
df.groupby(['Facility Name']).sum().head()
```

```python
df.groupby('Facility Name').sum().reset_index().head()
```

```python
df.groupby('Facility Name').sum()['Total_Cost'].head()
```

```python
df.groupby('Facility Name').sum().iloc[:,-1].head()
```

```python
df[['Facility Name', 'Total_Cost']].groupby('Facility Name').sum().reset_index().head()
```
